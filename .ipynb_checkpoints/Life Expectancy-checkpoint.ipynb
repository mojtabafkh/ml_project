```python
import pandas as pd
import numpy as np

Df = pd.read_csv("https://raw.githubusercontent.com/EvidenceN/DataSets/master/Life%20Expectancy%20Data.csv")

Df.info()
Df.head()
Df.describe()
```

```python
Df.columns = Df.columns.str.strip()

numeric_cols = Df.select_dtypes(include=[np.number]).columns

Df[numeric_cols] = Df.groupby('Country')[numeric_cols].transform(lambda x: x.fillna(x.mean()))

Df[numeric_cols] = Df[numeric_cols].fillna(Df[numeric_cols].mean())

Df_model = Df.drop(columns=['Country'])

print(Df_model.isnull().sum().sum())
```

```python
print(Df.isnull().sum())

numeric_cols = Df.select_dtypes(include=[np.number]).columns

Df[numeric_cols] = Df.groupby('Country')[numeric_cols].transform(lambda x: x.fillna(x.mean()))

Df[numeric_cols] = Df[numeric_cols].fillna(Df[numeric_cols].mean())
```

```python
Df['Status'] = Df['Status'].map({'Developed': 1, 'Developing': 0})

Df_model = Df.drop(columns=['Country'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
sns.heatmap(Df_model.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


Df_model.columns = Df_model.columns.str.strip()

X = Df_model.drop(columns=['Life expectancy'])
Y = Df_model['Life expectancy']

numeric_cols = Df_model.select_dtypes(include=[np.number]).columns
Df_model[numeric_cols] = Df_model.groupby('Country')[numeric_cols].transform(lambda x: x.fillna(x.mean()))

Df_model[numeric_cols] = Df_model[numeric_cols].fillna(Df_model[numeric_cols].mean())
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=33)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```
